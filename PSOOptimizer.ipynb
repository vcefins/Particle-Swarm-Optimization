```python
from fn_utils import *
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display
import time
import matplotlib as mpl
from PSO import *
mpl.rcParams['figure.dpi'] = 300
```

```python
# Simple Convex Function

PSO = SwarmOptimizer(n_particles=20, v_coeff=0.9, p_coeff=2, g_coeff=2)
a, b = 0, 100
function = function_classes.SimpleConvex()
PSO.set_objective_function(function)

fig = plt.figure()
X, Y, Z = F('simple_convex')
p = plt.pcolor(X, Y, Z)
fig.colorbar(p)
x_min = np.min(Z, axis=0)
y_min = np.min(Z, axis=1)
plt.scatter(0, 0, color='white')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

for i in range(20):
    PSO.step()
    for particle in PSO.get_particles():
        plt.scatter(*particle)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    

print("Particle position: ", PSO.get_population())
print("Global minimum at: ({},{})".format(0, 0))
print("Value at global minimum: ", function.get_value(0, 0))
```

```python
PSO = SwarmOptimizer(n_particles=20, v_coeff=0.2, p_coeff=1, g_coeff=0.07)
a, b = 0, 100
function = function_classes.Rosenbrock(a, b)
PSO.set_objective_function(function)

fig = plt.figure()
X, Y, Z = F('rosenbrock', rosen_a=a, rosen_b=b)
p = plt.pcolor(X, Y, Z)
fig.colorbar(p)
x_min = np.min(Z, axis=0)
y_min = np.min(Z, axis=1)
plt.scatter(a, a**2, color='white')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

for i in range(50):
    PSO.step()
    for particle in PSO.get_particles():
        plt.scatter(*particle)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    
population = PSO.get_population()
print("Particle position: ", population)
print("Global minimum at: ({},{})".format(a,a**2))
print("Value at global minimum: ", function.get_value(a,a**2))
```
