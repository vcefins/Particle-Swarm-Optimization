```python
from utils import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from optimizer import Optimizer
opt = Optimizer()
particle_coordinates = np.array(opt.initialize_particles())
```

```python
fig = plt.figure() 
ax = fig.gca(projection='3d') 
X, Y, Z = F('rastrigin', 10, 500)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
  cmap=cm.nipy_spectral, linewidth=0.08,
  antialiased=True)   

# plt.savefig('rastrigin_graph.png')
plt.show()
```

```python
fig = plt.figure() 
X, Y, Z = F('rastrigin', 10, 500)
plt.contourf(X, Y, Z)

plt.show()
```

```python
fig = plt.figure() 
ax = fig.gca(projection='3d') 
X, Y, Z = F('simple_convex', 10, 100)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
  cmap=cm.nipy_spectral, linewidth=0.08,
  antialiased=True)    

plt.show()
```
